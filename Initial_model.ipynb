```python
# Dropped all the duplicated values
df.drop_duplicates(subset =["Book-Title"], inplace=True)
```

```python
# This line of code will remove some unwanted data that are include inside () words after : , which does not hold much information
df["Book-Title"]=df["Book-Title"].str.split(r'[./(:]').str[0]
```

```python
wanted_columns = ["ISBN","Book-Title","Book-Author", "Year-Of-Publication","Publisher", "Genre" , "Description"]
```

```python
df.head()
```

```python
# This removes the rows and columns that contains the null value
df_1 = df[wanted_columns]
df_1.dropna(inplace=True)
df_1.head()
```

```python
# This will give the ISBN number of each of the unique books
books_we_have = df_1["ISBN"].unique()
books_we_have
```

```python
df_1 = df_1.drop_duplicates(subset =["Book-Title"])
df_1 = df_1.reset_index(drop=True)
```

```python
df_1["Description"][22]
```

```python
df_1["Combined_feature"] = df_1["Book-Title"] + " " + df_1["Book-Author"] + " " + df_1["Publisher"] + " " + df_1["Genre"] + " " +df_1["Description"]
```

```python
df_1.head(10)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# Initialize the tfidf
tfidf = TfidfVectorizer(stop_words="english")
tfidf_matrix = tfidf.fit_transform(df_1["Combined_feature"])
```

```python
tfidf_matrix.shape
```

```python
sparse_size = (
    tfidf_matrix.data.nbytes +
    tfidf_matrix.indices.nbytes +
    tfidf_matrix.indptr.nbytes
)

print(f"Actual Sparse matrix memory: {sparse_size / (1024 ** 2):.2f} MB")
```

```python
import sys
print(f"Dense matrix size: {(sys.getsizeof(tfidf_matrix.toarray()))/(1024 **3):.2f} GB")
print(f"There is massive compression that is 99.85% reduction on in the size")
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Compute pairwise similarity martix
cosine_sin = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
indices = pd.Series(df_1.index,index= df_1["Book-Title"]).drop_duplicates()


def recommend_books(title, cosine_sim=cosine_sin, top_n=5):
    # Get the index of the book that matches the title
    idx = indices[title]

    # Get the pairwise similarity scores for that book
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the books based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the top_n most similar books (excluding itself)
    sim_scores = sim_scores[1:top_n+1]

    # Get the book indices
    book_indices = [i[0] for i in sim_scores]

    # Return the top recommended books
    return df_1['Book-Title'].iloc[book_indices]
```

```python
recommend_books("High Tide in Tucson")
```
